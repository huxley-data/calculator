```python
obj_col = []
for i,j in zip(df.dtypes.index, df.dtypes.values) :
    if j == 'object' :
        obj_col.append(i)
```

```python
obj_col
```

```python
bool_col = []
for i,j in zip(df.dtypes.index, df.dtypes.values) :
    if j == 'bool' :
        bool_col.append(i)
```
